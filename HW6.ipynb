```python
import requests
import json
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
from pprint import pprint
from config import api_key
```

```python
#save config information
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

#build my query URL with the varaibles
query_url = f"{url}appid={api_key}&units={units}&q="


```

```python
cities = []
humidity = []
clouds = []
windspeed = []
lat = []
temp = []
long = []
date = []
country = []
valid_cities = []
```

```python
#open and read in CSV File, append each city to cities list.
with open("cities.csv") as cityfile:
    cityData = csv.reader(cityfile)
    for city in cityData:
        cities.append(city[0])
```

```python
for city in cities:
     try:
         response = requests.get(query_url + city)
         print(response.url)
         weather_json = response.json()
         lat.append(weather_json['coord']['lat'])
         long.append(weather_json['coord']['lon'])
         temp.append(weather_json['main']['temp'])
         clouds.append(weather_json['clouds']['all'])
         humidity.append(weather_json['main']['humidity'])
         windspeed.append(weather_json['wind']['speed'])
         date.append(weather_json['dt'])
         country.append(weather_json['sys']['country'])
         valid_cities.append(weather_json['name'])
     except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```

```python
#create a dataframe for cities, humidity, clouds, windspeed, lat, temp
city_data_pd = pd.DataFrame({
    "City":valid_cities,
    "Cloudiness":clouds,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "Latitude": lat,
    "Longitude": long,
    "Max Temp": temp,
    "Wind Speed":windspeed    
})
city_data_pd.head()
```

```python
print(len(valid_cities))
print(len(windspeed))
print(len(humidity)) 
print(len(clouds))
print(len(lat))
print(len(temp))
print(len(long))
print(len(date))
print(len(country))
```

```python
city_data_pd.plot(kind="scatter", x="Latitude", y="Max Temp", grid=True,figsize=(9,6),
              title="City Latitude vs. Max Temperature (04/01/2018)", edgecolors="black", alpha=1)
plt.savefig("LatitudevsTemperature.png")
plt.show()
```
